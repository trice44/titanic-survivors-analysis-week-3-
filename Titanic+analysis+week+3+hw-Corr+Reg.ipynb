```python
#week 3
#load data
import pandas as pd

df = pd.read_excel("titanic3.xls")

df.head()
```

```python
#see total number of passengers and assess missing data

df.count()
```

```python
#drop columns that are less relevant for analysis such as boat, body, home, ticket
df.drop(['cabin','boat','body','home.dest','ticket'], axis=1, inplace=True)
df.head()
```

```python
df.count()
```

```python
#convert city embarked into numerical value

def stringtonum(x):
    if x=='C':
        return 1
    if x=='Q':
        return 2
    if x=='S':
        return 3
    
```

```python
#create embark numerical

df['embark_value'] = df['embarked'].apply(stringtonum) 
df.head()
```

```python
#create gender numerical

def gendernum(x):
    if x=='female':
        return 1
    if x=='male':
        return 0
```

```python
#create gender numerical

df['gender_value'] = df['sex'].apply(gendernum) 
df.head()
```

```python
#drop rows with missing data in the variables of interest
df_no_missing = df.dropna()
df_no_missing.count()
```

```python
df_no_missing.head()
```

```python
#look at averages in the survivors and non-survivors
df_no_missing.groupby('survived').mean()
```

```python
#pivot table
pd.pivot_table(df_no_missing, index=['sex'])
```

```python
#pivot table by gender and class of avg survival
pd.pivot_table(df_no_missing, index=['sex','pclass'], values=['survived'])
```

```python
#pivot table by gender and presence of parent or child of avg survival
pd.pivot_table(df_no_missing, index=['sex','parch'], values=['survived'])
```

```python
df_no_missing.corr()
```

```python
#regression with intercept
import statsmodels.formula.api as sm
result = sm.ols(formula='survived ~ age + pclass + sibsp + parch + embark_value + gender_value', data=df_no_missing).fit()
result.summary()
```

```python
#regression without intercept
import statsmodels.formula.api as sm
result = sm.ols(formula='survived ~ age + pclass + sibsp + parch + embark_value + gender_value - 1', data=df_no_missing).fit()
result.summary()
```

```python
#regression with intercept on gender, the variables with strongest correlation
#regression with intercept
import statsmodels.formula.api as sm
result = sm.ols(formula='survived ~ gender_value', data=df_no_missing).fit()
result.summary()
```

```python
#regression without intercept on gender, the variables with strongest correlation
#regression with intercept
import statsmodels.formula.api as sm
result = sm.ols(formula='survived ~ gender - 1 ', data=df_no_missing).fit()
result.summary()
```

```python
#regression without intercept on gender, the variables with strongest correlation
#regression with intercept
import statsmodels.formula.api as sm
result = sm.ols(formula='survived ~ gender_value - 1 ', data=df_no_missing).fit()
result.summary()
```

```python
#regression without intercept on gender value and cabin class, 
import statsmodels.formula.api as sm
result = sm.ols(formula='survived ~ gender_value + pclass - 1', data=df_no_missing).fit()
result.summary()
```

```python
#regression without intercept on embarked city and cabin class, the variables with strongest correlation
#regression with intercept
import statsmodels.formula.api as sm
result = sm.ols(formula='survived ~ pclass + embark_value - 1', data=df_no_missing).fit()
result.summary()
```

```python
#regression without intercept on age
import statsmodels.formula.api as sm
result = sm.ols(formula='survived ~ age - 1', data=df_no_missing).fit()
result.summary()
```
